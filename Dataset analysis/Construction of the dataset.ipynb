For each of the 132 molecules of the Head-Gordon (HG) database we need to specify if the molecule can be analyzed using a spin-polarized (SP) or a non spin-polarized (NSP) analysis. Moreover we associate to each molecule several couples of the form (xc-psp). The object : molecule type + (xc-psp) represent a study and the statical polarizabiliry of each study will be computed through the appropriate workflow
